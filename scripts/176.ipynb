```python
def get_divs(n):
    divs = [[] for _ in range(n + 1)]
    for i in range(1, n + 1):
        for q in range(i, n + 1, i):
            divs[q].append(i)
    return divs
```

```python
from random import randint
def isprime(n, k=5):
    if n <= 1:
        return False
    if n == 2 or n == 3:
        return True
    d = n - 1
    s = 0
    while d % 2 == 0:
        d //= 2
        s += 1
    
    y = 0
    for _ in range(k):
        a = randint(2, n - 2)
        x = pow(a, d, n)
        for _ in range(s):
            y = pow(x, 2, n)
            if y == 1 and x != 1 and x != n - 1:
                return False
            x = y
        if y != 1:
            return False
    return True
```

```python
def sieve(n):
    s = [1 for _ in range(n + 1)]
    primes = []
    for q in range(2, n + 1):
        if s[q]:
            primes.append(q)
            for i in range(2 * q, n + 1, q):
                s[i] = 0
    return primes
primes = sieve(100_000)
def prime(i):
    return primes[i]
```

```python
def prod(k):
    t = 1
    for e in k:
        t *= e
    return t
```

```python
from math import prod
from sympy import isprime, divisors, prime
def A038547(n):
    def mult_factors(n):
        if isprime(n):
            return [(n, )]
        c = []
        for d in divisors(n, generator=True):
            if 1<d<n:
                for a in mult_factors(n//d):
                    c.append(tuple(sorted((d, )+a)))
        return list(set(c))
    return min((prod(prime(i)**(j-1) for i, j in enumerate(reversed(d), 2)) for d in mult_factors(n)), default=1) # 
```

```python
from collections import Counter
def dsq_cnt(n):
    pf = pfs(n)
    c = Counter(pf)
    t = 1
    for e in c.values():
        t *= (e * 2 + 1)
    return t
```

```python
i = 1
ma = 0
while True:
    c = dsq_cnt(i)
    if c == k * 2 + 1:
        print(i)
        break
    if c > ma:
        ma = c
        # print("NMA", i, c)
        print(i)
    i += 1
```

```python
import sys

def main():
    # Set limit from command-line argument, defaulting to 300
    limit = 3000
    maxsol = 0
    n = 2 * 3 * 5  # start at 30
    last = []

    while True:
        numsol = 0
        # x ranges from n+1 to 2*n inclusive
        for x in range(n + 1, n + n + 1):
            # Check if the division is exact: (x * n) should be divisible by (x - n)
            if (x * n) % (x - n) == 0:
                # y is the exact result of the division
                y = (x * n) // (x - n)
                numsol += 1

        if numsol > maxsol:
            last.append(n)
            print(f"New max n={n} sols={numsol},dsq-{dsq_cnt(n)}", end="\n")
            maxsol = numsol

        if len(last) > limit:
            break

        G = 30
        if len(last) > 2:
            G = gcd(*last[-2:])
        n += G # Increment n by 30
main()
```
