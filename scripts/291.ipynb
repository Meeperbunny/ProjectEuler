```python
from tqdm.notebook import trange
```

```python
s = {}
for x in trange(1, 10000):
    for y in range(1, x):
        n, d = (x**4-y**4), (x**3 + y**3)
        if n % d == 0:
            if n * d > 0:
                num = n // d
                # print(x, y, num)
                if num not in s:
                    s[num] = []
                s[num].append((x, y))
```

```python
from fractions import Fraction
v_f = {}
for e in s:
    print(e)
    for q in s[e]:
        F = Fraction(q[0], q[1])
        if F not in v_f:
            v_f[F] = (q[0], q[1])
        if q[0] < v_f[F][0]:
            v_f[F] = (q[0], q[1])
        x,y =q
        
        print(q, (x**4-y**4)// (x**3 + y**3), end=" ")
    print()
```
