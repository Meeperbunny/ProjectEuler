```python
def mex(s):
    for i in range(0, len(s) + 1):
        if i not in s:
            return i
```

```python
def G(N):
    nimbers = [[0 for _ in range(N)] for _ in range(N)]
    for i in range(0, N):
        for q in range(0, N):
            s = set()
            for p in [2, 3, 5, 7]:
                if i - p >= 0:
                    s.add(nimbers[i - p][q])
                if q - p >= 0:
                    s.add(nimbers[i][q - p])
            nimbers[i][q] = mex(s)
    return nimbers
```

```python
from collections import Counter
```

```python
def get_cnts(n):
    t = G(9)
    # print_aligned(t)
    x, y = n // 9, n // 9
    over = n % 9
    cnts = Counter()
    s_cnts = Counter()
    for v in t:
        for e in v:
            cnts[e] += x * y
    for i in range(over):
        for q in range(9):
            s_cnts[t[i][q]] += 1
    for e in s_cnts:
        cnts[e] += s_cnts[e] * 2 * x
    for i in range(over):
        for q in range(over):
            cnts[t[i][q]] += 1
    return cnts
```

```python
def xor_pow(cnter, k):
    b_c = Counter({0: 1})
    for i in range(k):
        new_cnts = Counter()
        for e in cnter:
            for q in b_c:
                new_cnts[e ^ q] += cnter[e] * b_c[q]
        b_c = new_cnts.copy()
    print(b_c)
    return b_c
```

```python
n, m = 10_000_019, 100
a = xor_pow(get_cnts(n), m)
sum(a[e] for e in a if e != 0) % 10**9
```
