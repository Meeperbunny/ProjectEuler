```python
def law_sines(a, b, c):
    inner = (a**2 + b**2 - c**2) / (2 * a * b)
    # print(inner)
    return degrees(acos(inner))
```

```python
from tqdm.notebook import trange
```

```python
N = 1000
T = 0
iaset = set()
ms = 0
for a in trange(1, N):
    for b in range(a, N):
        for c in range(b, a + b):
            if a + b + c > N:
                break
            # print(a, b, c)
            A, B, C = law_sines(b, c, a), law_sines(c, a, b), law_sines(a, b, c)
            # print(a, b, c, A, B, C)
            if a != b:
                if sum(isclose(e, round(e), abs_tol=1e-9) for e in {A, B, C}) >= 1:
                    ms = max(sum(isclose(e, round(e), abs_tol=1e-9) for e in {A, B, C}), ms)
                    print(a, b, c, A, B, C)
                    T += 1
                if len(iaset | {e for e in {A, B, C} if isclose(e, round(e), abs_tol=1e-9)}) > len(iaset):
                    print(a, b, c, A, B, C)
                iaset |= {e for e in {A, B, C} if isclose(e, round(e), abs_tol=1e-9)}
```
