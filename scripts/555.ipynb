```python
def M(m, k, s, n):
    return n - s if n > m else M(m, k, s, M(m, k, s, n + k))
```

```python
def SF(m, k, s):
    T = 0
    for i in range(1, 1000):
        if M(m, k, s, i) == i:
            print(m, k, s, i)
            T += i
    return T
            

def S(p, m):
    T = 0
    for s in range(1, p):
        for k in range(s + 1, p + 1):
            T += SF(m, k, s)
    return T
```

```python
def get_divs(n):
    divs = [[] for _ in range(n + 1)]
    for i in range(1, n + 1):
        for q in range(i, n + 1, i):
            divs[q].append(i)
    return divs
D = get_divs(10**6)
```

```python
def tri(n):
    return (n * (n + 1)) // 2
```

```python
T = 0
M = 10**6
for s in trange(1, M + 1):
    C = 0
    for d in D[s]:
        st = s + d
        if st <= M:
            C += d * (M - s) + tri(d)
    # print(s, C)
    T += C
T
```
