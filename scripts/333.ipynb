```python
s = set()
for i in range(0, 21):
    for q in range(0, 20):
        if 2**i * 3**q > N:
            continue
        else:
            s.add(2**i * 3**q)
len(s)
```

```python
nodiv = {}
for e in s:
    nodiv[e] = set()
    for o_e in s:
        if o_e % e == 0 or e % o_e == 0:
            pass
        else:
            nodiv[e].add(o_e)
```
