```python
def seq(n):
    s = [0 for _ in range(0, n + 1)]
    s[1] = 1
    for i in range(2, n + 1):
        if i % 2 == 0:
            t = i // 2
            s[i] = 2 * s[t]
        else:
            t = (i - 1) // 2
            s[i] = s[t] - 3 * s[t + 1]
    return s[1:]
```

```python
ra = 1, 10**12
r = ra
def t(n):
    if n % 2 == 0:
        return [n // 2]
    else:
        return [(n - 1) // 2, (n + 1) // 2]

def reduce_range(ra):
    return (
        min(*t(ra[0]), *t(ra[1])),
        max(*t(ra[0]), *t(ra[1]))
    )
```
