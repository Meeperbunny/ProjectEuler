```python
from copy import deepcopy
```

```python
def get_dist(n, k):
    b = [0 for _ in range(n * k + 1)]
    b[0] = 1
    for d in range(n):
        nxt = [0 for _ in range(n * k + 1)]
        for i in range(0, k * n + 1):
            # Mult through
            if b[i]:
                for z in range(1, k + 1):
                    nxt[i + z] += b[i] * (1 / k)
        b = deepcopy(nxt)
    return b
```
