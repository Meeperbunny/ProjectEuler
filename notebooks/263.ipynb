```python
def get_divs(n):
    divs = [[] for _ in range(n + 1)]
    for i in range(1, n + 1):
        for q in range(i, n + 1, i):
            divs[q].append(i)
    return divs
def sieve(n):
    s = [1 for _ in range(n + 1)]
    primes = []
    for q in range(2, n + 1):
        if s[q]:
            primes.append(q)
            for i in range(2 * q, n + 1, q):
                s[i] = 0
    return primes
M = 1_00000_000
primes = sieve(M)
# D = get_divs(M)
p_dict = {}
for i in range(len(primes)):
    p_dict[primes[i]] = i
```

```python
def pfs(n):
    pfs = []
    for i in range(2, n):
        if i * i > n:
            break
        while n % i == 0:
            n //= i
            pfs.append(i)
    if n - 1:
        pfs.append(n)
    return pfs
```

```python
def detect_cycle(l: list, k=3):
    for c_len in range(1, len(l)):
        isg = True
        for z in range(1, k):
            if l[0:c_len] != l[c_len * z: c_len * (z + 1)]:
                isg = False
                break
        if isg:
            return c_len
    return -1
detect_cycle([1, 2, 3, 1, 2, 3, 7, 1, 2, 3, 1, 2, 3, 7, 1, 2, 3, 1, 2, 3, 7, 1])
```

```python
l = [isgood(i) for i in range(2, 10_000, 2)]
detect_cycle(l[20:], k=5)
```
