```python
n = 100
cbrt = int(round(n**(1/3)))
sieve = [1 for _ in range(cbrt + 1)]
primes = []
for q in range(2, cbrt + 1):
    if q * q * q > n:
        break
    if sieve[q]:
        primes.append(q)
        for i in range(2 * q, cbrt + 1, q):
            sieve[i] = 0
```

```python
%%time
s = SortedSet([(1, ())])
for p in primes:
    # print(len(s))
    n_s = set()
    for v, p_s in s:
        if v * p * p * p > n:
            break
        n_s.add((v * p * p * p, tuple(list(p_s) + [p])))
    for e in n_s:
        s.add(e)
```
