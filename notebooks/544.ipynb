```python
import numbers
class Polynomial:
    def __init__(self, args):
        if isinstance(args, numbers.Number):
            self.coefs = list([args])
            return
        if len(args) == 0:
            self.coefs = []
            return
        if len(args) == 1 and isinstance(args, (list, tuple)) and isinstance(args[0], (list, tuple)):
            args = args[0]
        for e in args:
            assert(isinstance(e, numbers.Number))
        self.coefs = list(args)
    def at(self, x):
        c_x = 1
        T = 0
        for c in self.coefs:
            T += c_x * c
            c_x *= x
        return T
    def atMod(self, x, mod):
        for e in self.coefs:
            assert(isinstance(e, int))
        c_x = 1
        T = 0
        for c in self.coefs:
            T += c_x * c
            c_x *= x
            T %= mod
            c_x %= mod
        return T
    def add(self, other):
        assert(isinstance(other, Polynomial))
        self.new_poly = [0] * max(len(other.coefs), len(self.coefs))
        for i, c in enumerate(other.coefs):
            self.new_poly[i] += c
        for i, c in enumerate(self.coefs):
            self.new_poly[i] += c
        return Polynomial(self.new_poly)
    
    def sub(self, other):
        assert(isinstance(other, Polynomial))
        self.new_poly = [0] * max(len(other.coefs), len(self.coefs))
        for i, c in enumerate(self.coefs):
            self.new_poly[i] += c
        for i, c in enumerate(other.coefs):
            self.new_poly[i] -= c
        return Polynomial(self.new_poly)
    
    def mult(self, other):
        assert(isinstance(other, Polynomial))
        self.new_coefs = [0] * (len(other.coefs) + len(self.coefs) - 1)
        for i, a in enumerate(other.coefs):
            for q, b in enumerate(self.coefs):
                self.new_coefs[i + q] += a * b
        return Polynomial(self.new_coefs)
```
