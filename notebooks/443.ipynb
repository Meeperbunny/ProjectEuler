```python
def pfs(n):
    pfs = []
    for i in range(2, n):
        if i * i > n:
            break
        while n % i == 0:
            n //= i
            pfs.append(i)
    if n - 1:
        pfs.append(n)
    return pfs
```

```python
from functools import cache
from math import gcd
@cache
def g(n):
    if n == 4:
        return 13
    G = gcd(g(n - 1), n)
    if G != 1:
        print("###", n, g(n - 1) + G, G)
    return g(n - 1) + G
```
