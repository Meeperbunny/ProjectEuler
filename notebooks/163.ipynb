```python
import math

def angle_at(vertex, p1, p2):
    """
    Compute the angle at 'vertex' formed by the line segments vertex->p1 and vertex->p2.
    Returns the angle in radians.
    """
    # Create vectors from the vertex to the other two points.
    v1 = (p1[0] - vertex[0], p1[1] - vertex[1])
    v2 = (p2[0] - vertex[0], p2[1] - vertex[1])
    
    # Compute the magnitudes of the vectors.
    mag1 = math.sqrt(v1[0]**2 + v1[1]**2)
    mag2 = math.sqrt(v2[0]**2 + v2[1]**2)
    
    # If either magnitude is zero, the angle is undefined (points coincide) -> degenerate.
    if mag1 == 0 or mag2 == 0:
        return None
    
    # Compute the cosine of the angle using the dot product.
    dot = v1[0]*v2[0] + v1[1]*v2[1]
    cos_angle = dot / (mag1 * mag2)
    
    # Clamp the cosine value to the valid range [-1, 1] to account for floating-point issues.
    cos_angle = max(-1, min(1, cos_angle))
    
    return math.acos(cos_angle)

def is_degenerate_triangle(a, b, c, tol=1e-6):
    """
    Check if the triangle defined by points a, b, c is degenerate.
    A triangle is degenerate if one of its angles is within tol of 0 or π (i.e. the points are collinear).
    Each point is expected as a tuple (x, y).
    """
    # Quick check: if any two points are the same, the triangle is degenerate.
    if a == b or b == c or a == c:
        return True

    # Compute the angles at each vertex.
    angles = []
    for vertex, p1, p2 in [(a, b, c), (b, a, c), (c, a, b)]:
        angle = angle_at(vertex, p1, p2)
        # If the angle is undefined (None), then two points coincide.
        if angle is None:
            return True
        angles.append(angle)
    
    # Check if any angle is nearly 0 or nearly π.
    for angle in angles:
        if angle < tol or abs(angle - math.pi) < tol:
            return True
            
    return False
```

```python
def are_collinear(p1, p2, p3):
    x1, y1 = p1
    x2, y2 = p2
    x3, y3 = p3
    area = x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)
    # print("COLINEAR:", p1, p2, p3, area)
    return area < 0.01
```
