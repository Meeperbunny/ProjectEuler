```python
import math
from tqdm import tqdm

N = 10**6

def get_rad(r1, r2):
    prod = r1 * r2
    q = math.isqrt(prod)
    if q * q == prod:
        t = r1 + r2 + 2 * q
        if t and prod % t == 0:
            return prod // t
    return 0

def sieve(n):
    s = [1] * (n + 1)
    primes = []
    for q in range(2, n + 1):
        if s[q]:
            primes.append(q)
            for i in range(2 * q, n + 1, q):
                s[i] = 0
    return primes

def moebius(n):
    primes = sieve(n)
    m = [1] * (n + 1)
    for p in primes:
        for i in range(p, n + 1, p):
            m[i] *= -1
        for i in range(p * p, n + 1, p * p):
            m[i] = 0
    return m

# Precompute Moebius values up to sqrt(N)
M = moebius(int(math.isqrt(N) + 1))
tri = lambda x: (x * x + x) // 2

def num_squarefree(n):
    tot = 0
    for i in range(1, int(math.isqrt(n)) + 1):
        ni = n // (i * i)
        tot += (i * i) * ni * M[i] * (ni + 1)
    return tot / 2

T = 0

# --- Handle degenerate case: a == b ---
# In your original code, the only degenerate allowed was when r1 = 4.
a = 2
r1 = a * a  # 4
r2 = 4      # forced degenerate: r2 = r1 = 4
if get_rad(r1, r2) == 1:  # which is 1^2; i.e. b = 1.
    S = r1 + r2 + 1
    n_sq = num_squarefree(N // max(r1, r2))
    if r1 == 4:
        n_sq = tri(N // 4)
    T += n_sq * S

# --- Now handle the nondegenerate cases: we require a > b >= 1 ---
# The derivation gives c = ab/(a - b), so that r2 = c^2.
# Then get_rad(a^2, c^2) = (ac/(a+c))^2.
# Setting ac/(a+c) = b forces c = ab/(a-b), with a - b dividing ab exactly.
for a in tqdm(range(2, int(math.isqrt(N)) + 1)):
    r1 = a * a
    if r1 > N:
        break
    for b in range(1, a):
        diff = a - b
        if (a * b) % diff != 0:
            continue
        c = (a * b) // diff
        r2 = c * c
        if r2 < r1 or r2 > N:
            continue
        r3 = b * b
        # Verify using get_rad:
        if get_rad(r1, r2) != r3:
            continue
        S = r1 + r2 + r3
        n_sq = num_squarefree(N // max(r1, r2))
        T += n_sq * S

print(T)
```
