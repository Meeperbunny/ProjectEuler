```python
def getd(md):
    r = []
    b = [e for e in integer_partition(md) if len(e) <= 8]
    for e in tqdm(b):
        e = e + [0 for _ in range(8 - len(e))]
        r += list(set(permutations(e)))
    return r
```

```python
def fsum(n):
    sn = str(n)
    return sum(facs[ord(c) - ord('0')] for c in sn)
```

```python
def dsum(n):
    sn = str(n)
    return sum(ord(c) for c in sn) - len(sn) * ord('0')
```

```python
msz = 150
g = 150
mnum = 10**100
G = [mnum for _ in range(g + 1)]
G[0] = 0
for sz in range(1, msz + 1):
    print("On size {}".format(sz))
    d = getd(sz)
    for e in tqdm(d):
        e = list(e + (0,))
        t = 0
        for i in range(1, 10):
            t += factorial(i) * e[i - 1]
        for q in range(msz):
            c = dsum(t)
            if c <= g:
                snum = 0
                for i in range(1, 10):
                    for q in range(e[i - 1]):
                        snum = snum * 10 + i
                if snum > 0:
                    G[c] = min(snum, G[c])
            t += factorial(9)
            e[-1] += 1
    print("UK", sum(e == mnum for e in G))
    print(sum(dsum(e) for e in G))
```

```python
def GS(T):
    s = set({(T, tuple(0 for _ in range(10)))})
    for e in range(1, 9):
        nf = facs[e + 1]
        SS = set()
        for be, T in s:
            for q in range(e + 1):
                diff = q * facs[e]
                if (be - diff) % nf == 0:
                    T = list(T)
                    T[e] += q
                    SS.add((be - diff, tuple(T)))
        s = SS.copy()
    SS = list(SS)
    for i in range(len(SS)):
        T = list(SS[i][1])
        T[9] += SS[i][0] // facs[9]
        SS[i] = (SS[i][0], T)
    return SS
```

```python
def gnn(n):
    # 299999999
    # Find first non nine digit. If there is none, add a 1 to start
    s = list(str(n))
    for i in range(len(s)):
        if s[i] != '9':
            s[i] = chr(ord(s[i]) + 1)
            return int("".join(s))
    return int("1" + "".join(s))
```

```python
def ntt(n):
    t = list(0 for _ in range(10))
    if n <= 0:
        return t
    while n:
        t[n % 10] += 1
        n //= 10
    return t
```

```python
def tadd(a, b):
    comb = list(a[i] + b[i] for i in range(10))
    c = 0
    for i in range(10):
        ma = i + 1
        comb[i] += c
        c = comb[i] // ma
        if i != 9:
            comb[i] %= ma
    return comb
```

```python
def asum(t):
    T = 0
    for i in range(0, 10):
        T += i * t[i]
    return T
```

```python
def ffsum(t):
    T = 0
    for i in range(0, 10):
        T += facs[i] * t[i]
    return T
```
