```python
N = 10**13
```

```python
fib = [1, 1]
while fib[-1] < N:
    fib.append(fib[-1] + fib[-2])
for e in fib:
    print(e, end=" ")
fib = sorted(list(set(fib)))
```

```python
from collections import Counter
```

```python
def get_counts(l):
    s = Counter()
    s[0] = 1
    for e in l:
        to_add = Counter()
        for o in s:
            to_add[e + o] += s[o]
        s += to_add
    return s
```

```python
SC = get_counts(S)
TC = get_counts(T)
# SC, TC
```

```python
next_max = 0
sz = sum(F)
ways = [0 for _ in range(sz + 1)]
ways[0] = 1
i = 0
for f in tqdm(sorted(list(set(F)))):
    i += 1
    next_max += f
    # print(i, f, next_max, "#" if next_max > 10**13 else "")
    for i in range(sz, f - 1, -1):
        ways[i] += ways[i - f]
    # print(ways)
# ways
```

```python
waypre = [0 for _ in range(len(ways) + 1)]
for i in range(len(ways)):
    waypre[i + 1] = waypre[i] + ways[i]
def waysum(r):
    print(len(waypre), r)
    return waypre[r + 1]
```

```python
ans = 0
for sc in tqdm(SC):
    for tc in TC:
        # How many times we add the shift from sc + tc
        times = SC[sc] * TC[tc]
        shift = sc + tc
        ub = max(0, min(len(ways), N - shift + 1))
        s = waypre[ub]
        # print(sc, tc, s)
        ans += s * times
ans
```
