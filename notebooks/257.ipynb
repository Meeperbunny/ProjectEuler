```python
N = 10**8
D = get_divs(N)
T = 0
for a in trange(1, N // 3 + 1):
    b, c = a, a
    if a + b + c > N:
        break
    T += 1


for a in trange(1, N // 3 + 1):
    if a * 3 > N:
        break
    sqdivs = combine_divs(D[a], D[a])

    # K = 2
    fdivs = combine_divs(D[2], sqdivs)
    for i in fdivs:
        j = (2 * a**2) // i
        if i <= j:
            b = a + i
            c = a + j
            if a + b + c <= N:
                if a <= b <= c and a + b + c <= N and not (a + b <= c):
                    if b*c == a * (a+b+c):
                        T += 1
    
    # K = 3
    sdivs = combine_divs(D[3], sqdivs)
    for d in sdivs:
        o = (3 * a**2) // d
        i = (d - a) // 2
        j = (o - a) // 2
        if i <= j:
            b = a + i
            c = a + j
            if a + b + c <= N:
                if a <= b <= c and a + b + c <= N and not (a + b <= c):
                    if 2*b*c == a * (a+b+c):
                        T += 1
T
            
```
