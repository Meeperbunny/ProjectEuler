```python
def sieve(n):
    s = [1 for _ in range(n + 1)]
    primes = []
    for q in range(2, n + 1):
        if s[q]:
            primes.append(q)
            for i in range(2 * q, n + 1, q):
                s[i] = 0
    return primes
```

```python
def pfs(n):
    pfs = []
    for i in range(2, n):
        if i * i > n:
            break
        while n % i == 0:
            n //= i
            pfs.append(i)
    if n - 1:
        pfs.append(n)
    return pfs
```
