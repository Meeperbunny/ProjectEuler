```python
from collections import Counter
```

```python
def isduo(n):
    c = Counter()
    for i in str(n):
        c[i] += 1
    return len(c) <= 2
```

```python
# if end is zero, then we see what options we have for the ones place, and get the min for each. Never try to use 0?
def find_one(n):
    assert(n % 10 == 0)
    loop = n // 10 % 10
    base = loop
    seen = set()
    cnt = 2
    while loop not in seen:
        seen.add(loop)
        loop = base * cnt
        loop %= 10
        cnt += 1
    print(seen)
find_one(390)
```
