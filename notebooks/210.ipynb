```python
from tqdm import trange
from math import ceil
def solve(n):
    B = 1 + 2 * n * (n + 1) - 2 * n - 1 - n**2 // 2
    epsilon = 0.00000001
    R = (2**0.5)*(n/4)/2
    for c in trange(1, n // 2):
        # print(c, R**2 - (R - 1/(2**0.5)*c)**2)
        y = (R**2 - (R - 1/(2**0.5)*c)**2)**0.5
        if c & 1:
            y += 1/2**0.5
        B += int(2 * ((y - epsilon) // (2**0.5)))
        # print(c, (y - epsilon) // (2**0.5))
    return B

print(solve(10**4))
```

```python
def S(r):
    pnts = []
    for x in range(-r, r + 1):
        for y in range(-r, r + 1):
            if abs(x) + abs(y) <= r:
                pnts.append((x, y))
    return pnts
```

```python
import math

def triangle_angles(a, b, c):
    def distance(p1, p2):
        return math.hypot(p1[0] - p2[0], p1[1] - p2[1])

    def angle(opposite, side1, side2):
        cos_theta = (side1**2 + side2**2 - opposite**2) / (2 * side1 * side2)
        if not -1 <= cos_theta <= 1:
            return None
        return math.degrees(math.acos(cos_theta))

    ab = distance(a, b)
    bc = distance(b, c)
    ca = distance(c, a)

    area = 0.5 * abs((b[0] - a[0]) * (c[1] - a[1]) - (c[0] - a[0]) * (b[1] - a[1]))
    if area == 0:
        return None

    A = round(angle(bc, ab, ca), 8)
    B = round(angle(ca, ab, bc), 8)
    C = round(angle(ab, bc, ca), 8)

    return [A, B, C]
```

```python
def A(r):
    points = S(r)
    C = 0
    R = (r // 4, r // 4)
    O = (0, 0)
    for p in points:
        ta = triangle_angles(R, O, p)
        if ta:
            if 90 < max(ta) < 180:
                # print(p) #, max(ta))
                C += 1
    return C
```
