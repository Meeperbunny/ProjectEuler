```python
def mul(a, b):
    t = 0
    for i in range(0, 60):
        if (a >> i) & 1:
            t ^= b << i
    return t
```

```python
s = set()
for i in range(2, 100):
    for q in range(2, 100):
        print(bin(mul(i, q)), bin(i), bin(q))
        s.add(mul(i, q))
for i in range(2, 100):
    if i not in s:
        print(i, bin(i))
```
