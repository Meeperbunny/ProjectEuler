```python
def g(n):
    a = [[0 for _ in range(n + 1)] for _ in range(n + 1)]
    # print(a)
    for i in range(1, n + 1):
        for q in range(1, n + 1):
            t = min(i, q)
            s = [0 for _ in range(n * n)]
            for j in range(1, t):
                t1, t2 = j, t - j
                # print(i, q, i - t1, q - t2)
                s[a[i - t1][q - t2]] = 1
                s[a[i - t2][q - t1]] = 1
            if i < q:
                # print(i, q, "G")
                s[a[i][q - i]] = 1
            if i > q:
                # print(i, q, "L")
                s[a[i - q][q]] = 1
            for j in range(0, len(s)):
                if not s[j]:
                    # print(i, q)
                    a[i][q] = j
                    break
    return a
```

```python
a = g(44)
for i in a:
    print(i)
```

```python
p = {}
p[0] = 1
i = 0
while 2**i <= n:
    p[i + 1] = p[i] * 2 + 2**i
    i += 1
```

```python
f = False
tot = 0
for i in range(100, -1, -1):
    if (n >> i) & 1:
        if not f:
            tot -= f
        f = True
        tot += p[i] * 2
tot -= n * 2
tot += 2
tot -= len(p)
tot
```
