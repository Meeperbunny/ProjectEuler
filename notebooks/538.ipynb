```python
def A(a, b, c, d):
    s = (a + b + c + d) / 2
    return ((s - a) * (s - b) * (s - c) * (s - d))**0.5
```

```python
L = 1
r = 2
for i in range(100):
    r = i / 100
    print(r, A(L, L * r, L * r, L * r), A(L * r, L * r, L * r, L * r))
```

```python
import math

# Returns the number of set bits in k.
def B(k):
    return k.bit_count()

# Computes U(n) = 2^(B(3*n)) + 3^(B(2*n)) + B(n+1)
def U(n):
    return (2 ** B(3 * n)) + (3 ** B(2 * n)) + B(n + 1)

# Returns the area calculated using Brahmagupta's formula for a cyclic quadrilateral.
def getArea(a, b, c, d):
    s = (a + b + c + d) / 2
    return math.sqrt((s - a) * (s - b) * (s - c) * (s - d))

# Main code:
# Print U(i) for i from 1 to 10
print(" ".join(str(U(i)) for i in range(1, 11)))

N = 150
ans = 0
maxes = []  # This will store all computed U(n) values

for i in range(1, N + 1):
    u = U(i)
    maxes.append(u)
    
    # Skip until we have at least 4 elements.
    if i < 4:
        continue
    
    bestArea = 0
    bestPer = 0
    
    # Create a sorted (descending) list from the current values.
    sorted_desc = sorted(maxes, reverse=True)
    
    # Get the top 4 values as initial sides.
    sides = sorted_desc[:4]
    
    # Iterate over up to 200 remaining values.
    remaining = sorted_desc[4:4+200]
    for q, val in enumerate(remaining):
        A = getArea(sides[0], sides[1], sides[2], sides[3])
        if A > bestArea:
            bestArea = A
            bestPer = sum(sides)
        sides[q % 4] = val
        
    ans += bestPer
    
    if i in [5, 10, 150]:
        print(bestArea, bestPer)

print(ans)
```
