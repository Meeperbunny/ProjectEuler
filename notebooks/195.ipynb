```python
def get_divs(n):
    divs = [[] for _ in range(n + 1)]
    for i in range(1, n + 1):
        for q in range(i, n + 1, i):
            divs[q].append(i)
    return divs
```

```python
N = 100
D = get_divs(N)
```

```python
def combine_divs(d1, d2):
    dset = set()
    for a in d1:
        for b in d2:
            dset.add(a * b)
    return list(dset)
```

```python
def semiperimeter(a, b, c):
    return (a + b + c) / 2
```

```python
def area(a, b, c):
    s = semiperimeter(a, b, c)
    return (s * (s - a) * (s - b) * (s - c))**0.5
```

```python
def incircle_radius(a, b, c):
    return area(a, b, c) / semiperimeter(a, b, c)
```

```python
from tqdm import tqdm
```

```python
N = 1053779
```

```python
cnt = 0
inc = set()
M = 4
D = get_divs(M * N)
for j in tqdm(range(1, M * N + 1)):
    divs = D[j]
    jsq = j * j
    divs_jsq = combine_divs(divs, divs)
    for D1 in divs_jsq:
        D2 = jsq // D1
        A = (D1 + 2 * j)
        B = (D2 + 2 * j)
        if A % 3 != 0 or B % 3 != 0:
            continue
        A //= 3
        B //= 3
        a, b, c = A, B, (A + B - j)
        # print(A, B, (A + B + j))
        # print(a, b, c)
        r = incircle_radius(a, b, c)
        t = tuple(sorted([a, b, c]))
        if r <= N and A != B and t not in inc:
            inc.add(t)
            cnt += 1
cnt
```
