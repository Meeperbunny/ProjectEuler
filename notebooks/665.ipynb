```python
m = 30
n = m + 1
grundy = [[0 for _ in range(n)] for _ in range(n)]
grundy[0][0] = 0
```

```python
def mex(a):
    for i in range(0, len(a)):
        if not a[i]:
            return i
```

```python
for i in range(n):
    for q in range(n):
        if i == 0 and q == 0:
            continue
        gr = [0 for _ in range(n * n)]
        for j in range(1, max(i, q) + 1):
            if j <= i:
                gr[grundy[i - j][q]] = 1
            if j <= q:
                gr[grundy[i][q - j]] =  1
            if j <= q and j <= i:
                gr[grundy[i - j][q - j]] =  1
            if j * 2 <= q and j <= i:
                gr[grundy[i - j][q - 2 * j]] =  1
            if j <= q and j * 2 <= i:
                gr[grundy[i - 2 * j][q - j]] =  1
        grundy[i][q] = mex(gr)
```

```python
tot = 0
for i in range(n):
    for q in range(n):
        if not grundy[i][q] and i <= q and i + q <= m:
            tot += i + q
        if not grundy[i][q]:
            print(f"ðŸ¤¡", end="")
        else:
            print(f" {int(grundy[i][q])} ", end="")
    print()
```
