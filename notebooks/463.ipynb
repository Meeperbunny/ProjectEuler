```python
def f(n):
    if n == 1:
        return 1
    if n == 3:
        return 3
    if n % 2 == 0:
        return f(n // 2)
    if n % 4 == 1:
        k = n // 4
        return 2 * f(2 * k + 1) - f(k)
    if n % 4 == 3:
        k = n // 4
        return 3 * f(2 * k + 1) - 2 * f(k)
```

```python
T = 0
for i in range(1, 100 + 1):
    print(i, f(i))
    T += f(i)
T
```
