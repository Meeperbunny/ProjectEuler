```python
def f(n):
    cnt = 1
    for i in range(1, n):
        c = n - i
        if (c & i) == i:
            cnt += 1
    return cnt
```

```python
# 1 -> [1, 1]
# 2 -> [2, 3]
# 3 -> [4, 7]
def base(level: int):
    return 2**(level - 1)
```

```python
def idx_to_num(idx: int, level: int):
    max_ind = 2**(level - 1)
    diff = max_ind - idx
    if idx & 1:
        return (diff - idx) // 2
    else:
        # 0, 2  2 -> 2, 0 -> 3
        print((max_ind - idx - 2) // 2)
        return (max_ind - idx - 2) // 2 + (max_ind // 2)
```
