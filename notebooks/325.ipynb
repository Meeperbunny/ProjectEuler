```python
def mex(s):
    for i in range(0, len(s)):
        if i not in s:
            return i
    return len(s)
```

```python
def G(n):
    gru = [[-1 for _ in range(n + 1)] for _ in range(n + 1)]
    for i in range(0, n + 1):
        for q in range(i + 1, n + 1):
            s = set()
            if i != 0:
                for d in range(i, q + 1, i):
                    s.add(gru[min(q - d, i)][max(q - d, i)])
            # print(i, q)
            gru[i][q] = mex(s)
    return gru
                
```

```python
def print_aligned(matrix):
    col_widths = [max(len(str(matrix[row][col])) for row in range(len(matrix))) 
                  for col in range(len(matrix[0]))]

    for row in matrix:
        print(" ".join(f"{num:>{col_widths[i]}}" for i, num in enumerate(row)))
```

```python
n = 20
```

```python
g = G(n)
t = 0
for i in range(1, n + 1):
    for q in range(i + 1, n + 1):
        if (g[i][q] == 0):
            t += i + q
print_aligned(g)
print(t)
```

```python
def sosq(n):
    return n * (n + 1) * (2 * n + 1) // 6
def lower(L, R):
    return L * tri(R) + R * tri(R - 1) - sosq(R - 1)
def upper(L, R):
    return L * tri(R) + tri(R - 1) + sosq(R - 1)
```
