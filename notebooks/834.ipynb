```python
def seq(n, k=10000):
    s = [n]
    for i in range(1, k + 1):
        s.append(s[-1] + n + i)
    return s[1:]
```

```python
def S(s):
    n = s[0] // 2
    ans = []
    for i in range(0, len(s)):
        if s[i] % (n + i + 1) == 0:
            ans.append(i + 1)
    return ans
```

```python
for n in range(3, 50 + 1):
    ss = S(seq(n))
    print(n, max(ss), sum(ss), ss)
```
