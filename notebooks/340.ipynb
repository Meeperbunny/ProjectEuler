```python
from functools import cache
# a, b, c = 21**7, 7**21, 12**7
a, b, c = 21, 49, 12
@cache
def F(n):
    global a, b, c
    if n > b:
        return n - c
    return F(a + F(a + F(a + F(a + n))))
```
