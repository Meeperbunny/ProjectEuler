```python
def mex(s: set):
    for i in range(0, len(s)):
        if i not in s:
            return i
    return len(s)
```

```python
def nim(N):
    is_loss = [[[-1 for _ in range(N + 1)] for _ in range(N + 1)] for _ in range(N + 1)]
    s = 0
    for a in range(0, N + 1):
        for b in range(a + 1, N + 1):
            for c in range(b + 1, N + 1):
                is_win = False
                for i in range(1, c + 1):
                    l = [sorted(list(e)) for e in [(a - i, b, c), (a, b - i, c), (a, b, c - i)] if min(e) >= 0]
                    l = [e for e in l if e[0] != e[1] and e[1] != e[2]]
                    for el in l:
                        if is_loss[el[0]][el[1]][el[2]]:
                            is_win = True
                is_loss[a][b][c] = not is_win
                if a and is_loss[a][b][c]:
                    s += a + b + c
                    # print(a, b, c)
    return is_loss, s
loss, s = nim(128)
s
```

```python
def populate(l):
    for a in range(0, len(l)):
        for b in range(a + 1, len(l)):
            for c in range(b + 1, len(l)):
                gt = l[a][b][c]
                # l[a][c][b] = gt
                # l[b][a][c] = gt
                # l[b][c][a] = gt
                # l[c][a][b] = gt
                # l[c][b][a] = gt
    for a in range(1, len(l)):
        for b in range(a + 1, len(l)):
            for c in range(b + 1, len(l)):
                if a == b or b == c or a == c:
                    l[a][b][c] = -1
                if l[a][b][c] == 0:
                    l[a][b][c] = "[#]"
    return l
```

```python
def compress(l):
    CC = 0
    r = [[0 for _ in l] for _ in l]
    for a in range(0, len(l)):
        for b in range(a + 1, len(l)):
            for c in range(b + 1, len(l)):
                if l[a][b][c] == "[#]":
                    r[b][c] += 1
                    CC += 1
    for a in range(0, len(l)):
            for b in range(0, len(l)):
                r[a][b] = "[{}]".format(r[a][b])
    return r, CC
```

```python
def grundy(N):
    is_loss = [[[-1 for _ in range(N + 1)] for _ in range(N + 1)] for _ in range(N + 1)]
    s = 0
    for a in range(0, N + 1):
        for b in range(a + 1, N + 1):
            for c in range(b + 1, N + 1):
                g_seen = set()
                for i in range(1, c + 1):
                    l = [sorted(list(e)) for e in [(a - i, b, c), (a, b - i, c), (a, b, c - i)] if min(e) >= 0]
                    l = [e for e in l if e[0] != e[1] and e[1] != e[2]]
                    for el in l:
                        g_seen.add(is_loss[el[0]][el[1]][el[2]])
                is_loss[a][b][c] = mex(g_seen)
                if a and is_loss[a][b][c] == 0:
                    s += a + b + c
    return is_loss, s
loss, s = grundy(35)
s
```

```python
def print_aligned(matrix):
    col_widths = [max(len(str(matrix[row][col])) for row in range(len(matrix))) 
                  for col in range(len(matrix[0]))]

    for row in matrix:
        print(" ".join(f"{num:>{col_widths[i]}}" for i, num in enumerate(row)))
```

```python
def g(x):
    return 3*2**(x-1)*(2**x-1)
```

```python
def A_CNT(N):
    low = 1
    x = 1
    next_gap = 6
    per = 4
    sz = 2
    start = 5
    T = 0
    while start <= N:
        gx = g(x)
        # print(x, gx, start, per)

        diff = N - start + 1
        # Firsts
        occ = (diff + per - 1) // per
        cont = occ * gx * sz
        endsat = start + per * (occ - 1) - 1 + sz
        # print("adding", occ, cont, endsat)
        over = endsat - N
        if over > 0:
            cont -= over * gx
        # print(over, cont)
        T += cont

        start += next_gap
        next_gap *= 2
        per *= 2
        sz *= 2
        x += 1
    return T
A_CNT(12)

# 5 11 23 47
# 6 12 24
```

```python
def bsum(s, e):
    adder = s * 2
    l_e = (e - s) // 2
    l_o = (e - s - 1) // 2
    return adder * (tri(l_e) + tri(l_o))
bsum(7, 13)
```
